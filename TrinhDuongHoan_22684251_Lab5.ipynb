**Bài 1**

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
X = np.array([[1,0],[1,1],[1,2],[1,3],[1,4]])
Y = np.array([10,8,7,5,2])
A = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)
print(f'Phương trình tuyến tính là : y = %.1f %.1fx '%(A[0],A[1]))
```

**Bài 2**

```python
def model(x,a0,a1,a2):
    return a0 + a1*x + a2 * x**2 
```

- a

```python
X = np.array([2,3,5,6])
Y = np.array([0,-10,-48,-76])
# X = np.hstack((np.ones_like(X),X,X**2))

# A = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(Y)
A,E = curve_fit(model,X,Y)
print(f'Phương trình tuyến tính là : y = %.1f + %.1fx + %.1fx^2 '%(A[0],A[1],A[2]))
```

```python
plt.scatter(X, Y, label='Dữ liệu')
plt.plot(X, model(X, *A), color='red', label='Phương trình hồi quy')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```
